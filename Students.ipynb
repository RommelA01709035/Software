```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import numpy as np
from random import randint, choice
import string
import sys

class Estudiante:
    def __init__(self, nombre, identificacion):
        self.nombre = nombre
        self.identificacion = identificacion
        self.calificaciones = {"Español": 0, "Inglés": 0, "Matemáticas": 0, "Física": 0}
        self.asistencia = {}

    def generar_calificaciones(self):
        # Generar calificaciones aleatorias entre 60 y 100 para cada materia
        for materia in self.calificaciones:
            self.calificaciones[materia] = randint(60, 100)

    def calcular_promedio(self):
        return sum(self.calificaciones.values()) / len(self.calificaciones)

    def __str__(self):
        return f"Estudiante: {self.nombre}, ID: {self.identificacion}, Calificaciones: {self.calificaciones}"

    def __repr__(self):
        return self.__str__()

def generar_objeto_estudiante():
    # Generar un nombre aleatorio para el estudiante
    nombre = ''.join(choice(string.ascii_lowercase) for _ in range(6))
    identificacion = ''.join(choice(string.digits) for _ in range(4))

    # Crear el objeto Estudiante
    estudiante = Estudiante(nombre, identificacion)
    estudiante.generar_calificaciones()

    return estudiante

def calcular_promedio_estudiantes(b):
    if len(estudiantes) < 1:
        print("No hay estudiantes generados.")
    else:
        promedio_text = ""
        for estudiante in estudiantes:
            promedio = estudiante.calcular_promedio()
            promedio_text += f"Estudiante: {estudiante.nombre}, Promedio de calificaciones: {promedio}\n"

        print("Promedio de Calificaciones:")
        print(promedio_text)

def generar_estudiante(b):
    estudiante = generar_objeto_estudiante()
    estudiantes.append(estudiante)
    print(f"Estudiante generado: {estudiante}")

def cerrar_programa(b):
    clear_output()
    sys.exit(0)

# Lista para almacenar los estudiantes
estudiantes = []

# Crear los botones
boton_generar = widgets.Button(description="Generar Estudiante")
boton_generar.on_click(generar_estudiante)

boton_promedio = widgets.Button(description="Calcular Promedio")
boton_promedio.on_click(calcular_promedio_estudiantes)

boton_cerrar = widgets.Button(description="Cerrar programa")
boton_cerrar.on_click(cerrar_programa)

# Mostrar los botones
display(boton_generar)
display(boton_promedio)
display(boton_cerrar)
```
